```python
#5.2 Flipping an Image

# Use shape to access the image’s row and column sizes
img = cv2.imread("lena.png")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
horizon_flip = np.zeros(img.shape,img.dtype)
vertical_flip = np.zeros(img.shape,img.dtype)
horizon_flip_gray = np.zeros(img.shape,img.dtype)
vertical_flip_gray = np.zeros(img.shape,img.dtype)

# Make the empty images has the same properties as filpped pixels 
empty_original = np.zeros(img.shape,img.dtype)
empty_gray = np.zeros(img.shape,img.dtype)

# Use for loops to loop through the image’s pixels,and place the pixels in their corresponding flipped positions
# Horizontal flip
for j in range(len(img[0])):
    for k in range(len(img[1])):
        empty_original[j,k] = img[j,-k-1]
        empty_gray[j,k] = gray[j,-k-1]
        horizon_flip[j,k] = empty_original[j,k]
        horizon_flip_gray[j,k] = empty_gray[j,k] 
# Vertical flip
for j in range(len(img[0])):
    for k in range(len(img[1])):
        empty_original[j,k]  = img[-j-1,k]
        empty_gray[j,k] = gray[-j-1,k]
        vertical_flip[j,k] = empty_original[j,k]
        vertical_flip_gray[j,k] = empty_gray[j,k] 

# Plot the results
fig, axis = plt.subplots(2,2)
axis[0,0].imshow(cv2.cvtColor(horizon_flip,cv2.COLOR_BGR2RGB))
axis[0,0].set_title("Lena Color Horizontal Flip")

axis[0,1].imshow(cv2.cvtColor(vertical_flip,cv2.COLOR_BGR2RGB))
axis[0,1].set_title("Lena Color Vertical Flip")

axis[1,0].imshow(horizon_flip_gray,cmap="gray")
axis[1,0].set_title("Lena Horizontal Flip")

axis[1,1].imshow(vertical_flip_gray,cmap="gray")
axis[1,1].set_title("Lena Vertical Flip")
plt.tight_layout() 
```

```python
#5.3 Negating Image Pixels

#Define negate
img = cv2.imread("lena.png")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
Negate_Gray = gray
Negate_Color = img

#Loop through each image pixel and negate each pixels based on its data type
for a in range(3):
    for b in range(512):
        for c in range(512):
            Negate_Color[b][c][a]= abs(img[b][c][a]-255)
            Negate_Gray[b][c] = abs(gray[b][c] - 255)
            
# Plot the results
fig, axis = plt.subplots(1,2)
axis[0].imshow(cv2.cvtColor(Negate_Color,cv2.COLOR_BGR2RGB))
axis[0].set_title("Lena Color Negated")

axis[1].imshow(cv2.cvtColor(Negate_Gray,cv2.COLOR_BGR2RGB))
axis[1].set_title("Lena Gray Negated")
```

```python
#5.4 Separating Color Channels

# Create three empty color images of the original color image’s shape and size for visualization
img = cv2.imread("lena.png")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
empty_Blue = np.zeros(img.shape, img.dtype)
empty_Green = np.zeros(img.shape, img.dtype)
empty_Red = np.zeros(img.shape, img.dtype)
empty_Merge = np.zeros(img.shape, img.dtype)

#Loop through color image pixels
for j in range(512):
    for k in range(512):
        empty_Blue[j][k] =  ((img[j][k])[0], 0, 0)
        empty_Green[j][k] =  (0, (img[j][k])[1], 0)
        empty_Red[j][k] =  (0, 0, (img[j][k])[2])
        empty_Merge[j][k] = empty_Blue[j][k] + empty_Green[j][k] + empty_Red[j][k]

#Plot the results
plt.subplot(2,3,1)
plt.imshow(cv2.cvtColor(empty_Blue, cv2.COLOR_BGR2RGB))
plt.title("Lena blue image")

plt.subplot(2,3,2)
plt.imshow(cv2.cvtColor(empty_Green, cv2.COLOR_BGR2RGB))
plt.title("Lena green image")

plt.subplot(2,3,3)
plt.imshow(cv2.cvtColor(empty_Red, cv2.COLOR_BGR2RGB))
plt.title("Lena red image")
plt.tight_layout() 
```

```python
#5.3 continue
#Plot the merged result
plt.subplot(2,3,5)
plt.imshow(cv2.cvtColor(empty_Merge, cv2.COLOR_BGR2RGB))
plt.title("Merged Lena image")
```
